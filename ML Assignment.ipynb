```python
# List of desired columns to extract day of the week
desired_columns = ['date_added','read_at','date_updated','started_at']

# Function to extract day of the week
def extract_day_of_week(timestamp):
    if pd.isnull(timestamp):  # Check for NaN values
        return None
    date_obj = datetime.strptime(timestamp, "%a %b %d %H:%M:%S %z %Y")
    return date_obj.strftime("%a")

# Apply the function to desired columns and create new columns
for column in desired_columns:
    df[f'{column}_day_of_week'] = df[column].apply(extract_day_of_week)
    
df.head()
```

```python
import numpy as np
df['date_is_weekend']=np.where((df['date_added_day_of_week']).isin(['Sunday','Saturday']),1,0)
df.head()
```

```python
mode=df['read_at_day_of_week'].mode()[0]
print(mode)
df['read_at_day_of_week']=df['read_at_day_of_week'].fillna(mode)
```

```python

object_columns = df.select_dtypes(include=['object']).columns
df[object_columns] = df[object_columns].apply(lambda x: x.str.lower())
```

```python
import re
def remove(text):
    pattern=re.compile('\n')
    return pattern.sub(r'',text)
```

```python
df['review_text']=df['review_text'].apply(remove)
```

```python
import string,time
string.punctuation
```

```python
exclude=string.punctuation
```

```python
def remove_punc(text):
    return text.translate(str.maketrans('','',exclude))   
```

```python
df['review_text']=df['review_text'].apply(remove_punc)
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
target_column = 'review_text'

# Set English stopwords from NLTK
stop_words = set(stopwords.words('english'))

# Function to remove stopwords
def remove_stopwords(text):
    text = ' '.join(word for word in text.split() if word.lower() not in stop_words)
    return text

# Remove stopwords from the specified column
df[target_column] = df[target_column].apply(remove_stopwords)

# Save the modified DataFrame back to a CSV file
df.to_csv('training.csv', index=False)
```
